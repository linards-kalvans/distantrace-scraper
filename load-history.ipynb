```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
import os

base_url = "https://distantrace.com"

login_url = "/lv/konts/login/"

login_page = requests.get(base_url + login_url, headers={"User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0"})
```

```python
csfr_token = BeautifulSoup(login_page.text, "html.parser").find("input", {"name": "csrfmiddlewaretoken"})["value"]

print(csfr_token)
```

```python
login_page.cookies.get_dict()
```

```python
login_data = {
    "csrfmiddlewaretoken": [csfr_token, csfr_token],
    "login": os.getenv("DR_LOGIN"),
    "password": os.getenv("DR_PASSWORD")
}

time.sleep(1)
login_response = requests.post(
    login_url,
    data=login_data,
    headers={
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0",
        "Referer": login_url
    },
    cookies=login_page.cookies
)

print(login_response.text)

```

```python
parsed_login_response = BeautifulSoup(login_response.text, "html.parser")
```

```python
parsed_login_response.find("div", {"id": "pills-active-events"}).find("a")["href"]
```

```python
active_events = [h5.find("a")["href"] for h5 in parsed_login_response.find("div", {"id": "pills-active-events"}).find_all("h5", {"class": "card-title"})]

completed_events = [h5.find("a")["href"] for h5 in parsed_login_response.find("div", {"id": "pills-completed-events"}).find_all("h5", {"class": "card-title"})]
```

```python
time.sleep(random.uniform(1, 10))
events_response = requests.get(
    base_url + completed_events[0] + "dalibnieki/",
    headers={
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0",
        "Referer": base_url + active_events[0]
    },
    cookies=login_response.cookies
)

print(events_response.text)
```

```python
events_parsed = BeautifulSoup(events_response.text, "html.parser")

participants = [[a["href"] for a in tr.find_all("a") if "dalibnieki" in a["href"]] for tr in events_parsed.find("div", {"class": "table-container"}).find("table").find("tbody").find_all("tr")]

# Flatten list
participants = [item for sublist in participants for item in sublist]

participants
```

```python
time.sleep(random.uniform(1, 10))
participants_response = requests.get(
    base_url + participants[0],
    headers={
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0",
        "Referer": base_url + active_events[0] + "dalibnieki/"
    },
    cookies=login_response.cookies
)
```

```python

participants_parsed = BeautifulSoup(participants_response.text, "html.parser")
name = participants_parsed.find("h3", {"class": "text-secondary"}).text
id = participants[0].split("/")[-2]

results = pd.DataFrame(
    [{
        "id": id,
        "name": name,
        "date": pd.to_datetime(tr[1], dayfirst=True).date(),
        "distance": float(tr[2].replace(",", ".")),
        "steps": int(tr[3].replace(",", ""))
    } for tr in [[td.text for td in tr.find_all("td")] for tr in participants_parsed.find("div", {"class": "table-container"}).find("table").find("tbody").find_all("tr")]]
)

results.head()
```

```python
next_page_url = [a["href"] for a in participants_parsed.find("nav", {"class": "pagination"}).find_all("a") if not a["href"].startswith("#")][-1]

time.sleep(random.uniform(1, 10))

next_page_response = requests.get(
    base_url + participants[0] + next_page_url,
    headers={
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0",
        "Referer": base_url + participants[0]
    },
    cookies=login_response.cookies
)

next_page_parsed = BeautifulSoup(next_page_response.text, "html.parser")
```

```python
next_page_data = pd.DataFrame(
    [{
        "id": id,
        "name": name,
        "date": pd.to_datetime(tr[1], dayfirst=True).date(),
        "distance": float(tr[2].replace(",", ".")),
        "steps": int(tr[3].replace(",", ""))
    } for tr in [[td.text for td in tr.find_all("td")] for tr in next_page_parsed.find("div", {"class": "table-container"}).find("table").find("tbody").find_all("tr")]]
)

pd.concat([results, next_page_data])
```

```python
# 1. Get all completed events
# 2. Get all participants for each event
# 3. Get all results for each participant

all_results = pd.DataFrame()
for event in completed_events:
    event_id = event.split("/")[-2]
    print("Processing event", event_id)
    time.sleep(random.uniform(1, 10))
    participants_response = requests.get(
        base_url + event + "dalibnieki/",
        headers={"User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0"},
        cookies=login_response.cookies
    )
    participants_parsed = BeautifulSoup(participants_response.text, "html.parser")
    event_name = participants_parsed.find("h1").text
    print("Event name:", event_name)
    participants = [a["href"] for a in participants_parsed.find("div", {"class": "table-container"}).find("table").find("tbody").find_all("a") if "dalibnieki" in a["href"]]
    for participant in participants:
        print("Processing participant", participant)
        time.sleep(random.uniform(1, 10))
        participant_response = requests.get(
            base_url + participant,
            headers={"User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0"},
            cookies=login_response.cookies
        )
        participant_parsed = BeautifulSoup(participant_response.text, "html.parser")
        name = participant_parsed.find("h3", {"class": "text-secondary"}).text
        print("Participant name:", name)
        id = participant.split("/")[-2]
        results = pd.DataFrame(
            [{
                "event_id": event_id,
                "event_name": event_name,
                "participant_id": id,
                "participant_name": name,
                "date": pd.to_datetime(tr[1], dayfirst=True).date(),
                "distance": float(tr[2].replace(",", ".")),
                "steps": int(tr[3].replace(",", ""))
            } for tr in [[td.text for td in tr.find_all("td")] for tr in participant_parsed.find("div", {"class": "table-container"}).find("table").find("tbody").find_all("tr")]]
        )
        results.head()
        next_page_url = [a["href"] for a in participant_parsed.find("nav", {"class": "pagination"}).find_all("a") if not a["href"].startswith("#")]
        if next_page_url:
            print("Next page URL:", next_page_url[-1])
            time.sleep(random.uniform(1, 10))
            next_page_response = requests.get(
                base_url + participant + next_page_url[-1],
                headers={"User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:131.0) Gecko/20100101 Firefox/131.0"},
                cookies=login_response.cookies
            )
            next_page_parsed = BeautifulSoup(next_page_response.text, "html.parser")
            next_page_data = pd.DataFrame(
                [{
                    "event_id": event_id,
                    "event_name": event_name,
                    "participant_id": id,
                    "participant_name": name,
                    "date": pd.to_datetime(tr[1], dayfirst=True).date(),
                    "distance": float(tr[2].replace(",", ".")),
                    "steps": int(tr[3].replace(",", ""))
                } for tr in [[td.text for td in tr.find_all("td")] for tr in next_page_parsed.find("div", {"class": "table-container"}).find("table").find("tbody").find_all("tr")]]
            )
            results = pd.concat([results, next_page_data])
        all_results = pd.concat([all_results, results])

all_results.head()
```
